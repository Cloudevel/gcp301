# *BigQuery ML*.

*BigQuery ML* permite utilizar diversos algoritmos y modelos de *Machine Learning* usando *SQL* estándar.

* https://support.google.com/analytics/answer/7586738?hl=en#zippy=%2Cin-this-article
* https://cloud.google.com/bigquery-ml/docs/tutorials

## Modelos soportados.

* [Regresión lineal](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-glm).
* [Regresión logística binaria](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-glm)
* [Regresión logística multiclase](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-glm)

* [Agrupaqmiento de K-medias](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-kmeans)
* [Factorización de matrices](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-matrix-factorization)
* [Series de tiempo](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-time-series)
* [Boosted Tree](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-boosted-tree)
* [Tablas AutoML](https://cloud.google.com/bigquery-ml/docs/reference/standard-sql/bigqueryml-syntax-create-automl)
